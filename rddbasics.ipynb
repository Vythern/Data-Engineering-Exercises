```python
from pyspark import SparkContext, SparkConf

# Part 1.  Square each number
sc = SparkContext('local[*]', 'RDDBasics2')


nums = sc.parallelize(range(1, 10))

squared = nums.map(lambda x: x * x)

num_str = nums.map(lambda x: "num_" + str(x))

sc.stop()
```

```python
# rddbasics task 3 filter transformation

from pyspark import SparkContext, SparkConf

sc = SparkContext('local[*]', 'RDDBasics3')

nums = sc.parallelize(range(1, 10))

evens = nums.filter(lambda x: x % 2 == 0)

greater_than_5 = nums.filter(lambda x: x > 5)

evens_greater = nums.filter(lambda x: x > 5 and x % 2 == 0)

sc.stop()
```

```python
from pyspark import SparkContext, SparkConf

sc = SparkContext('local[*]', 'RDDBasics4')

sentences = sc.parallelize([
    "Hello World",
    "Apache Spark is Fast",
    "PySpark is Python plus Spark"
])

words = sentences.flatMap(lambda x: x.split())

word_lengths = words.map(lambda x: x + str(len(x)))

sc.stop()
```

```python
from pyspark import SparkContext, SparkConf

logs = sc.parallelize([
    "INFO: User logged in",
    "ERROR: Connection failed",
    "INFO: Data processed",
    "ERROR: Timeout occurred",
    "DEBUG: Cache hit"
])

errors = logs.filter(lambda x: x[0:5:1] == "ERROR")

capital_errors = errors.map(lambda x: x.upper())

# all words from all error messages
error_words = capital_errors.flatMap(lambda x: x.split())

sc.stop()
```
