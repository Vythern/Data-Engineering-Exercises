```python
from pyspark import SparkContext, SparkConf

# exercise_transformations
sc = SparkContext.getOrCreate(SparkConf().setMaster("local[*]").setAppName("Transformations"))

# Sample log data
logs = sc.parallelize([
    "2024-01-15 10:00:00 INFO User login: alice",
    "2024-01-15 10:01:00 ERROR Database connection failed",
    "2024-01-15 10:02:00 INFO User login: bob",
    "2024-01-15 10:03:00 WARN Memory usage high",
    "2024-01-15 10:04:00 ERROR Timeout occurred",
    "2024-01-15 10:05:00 INFO Data processed: 1000 records",
    "2024-01-15 10:06:00 DEBUG Cache hit rate: 95%"
])

errors = logs.map(lambda x: x.split()).filter(lambda x: x[2] == "ERROR")

levels = logs.map(lambda x: x.split()).map(lambda x: x[2])

error_messages = errors.map(lambda x: x[3:len(x):1], desc=true)

print(f"Get all error logs:  {errors.collect()}")
print(f"Only get the log level:  {levels.collect()}")
print(f"Messages only:  {error_messages.collect()}")

sc.stop()
```

```python
from pyspark import SparkContext, SparkConf

# exercise_transformations
sc = SparkContext.getOrCreate(SparkConf().setMaster("local[*]").setAppName("Transformations"))

words = sc.parallelize([
    "spark", "hadoop", "spark", "data", "hadoop", 
    "spark", "python", "data", "spark", "scala"
])

unique_words = words.distinct()

word_counts = words.map(lambda k: (k, 1)).reduceByKey(lambda a, b: a+b)

sorted_counts = word_counts.sortBy(lambda x: x[1], ascending=False)

print(f"Distinct Words:  {unique_words.collect()}")
print(f"Word counts: {word_counts.collect()}")
print(f"Words sorted by count: {sorted_counts.collect()}")

sc.stop()
```

```python
from pyspark import SparkContext, SparkConf

# exercise_transformations
sc = SparkContext.getOrCreate(SparkConf().setMaster("local[*]").setAppName("Transformations"))


set1 = sc.parallelize([1, 2, 3, 4, 5])
set2 = sc.parallelize([4, 5, 6, 7, 8])

combined = set1.union(set2)

intersect = set1.intersection(set2)

difference = set1.subtract(set2)

non_intersect = combined.subtract(intersect)

print(f"Union:  {combined.collect()}")
print(f"Intersect:  {intersect.collect()}")
print(f"Difference:  {difference.collect()}")
# Was curious about the non-intersect...  
print(f"Symmetric Difference:  {non_intersect.collect()}")



sc.stop()
```

```python
from pyspark import SparkContext, SparkConf

# exercise_transformations
sc = SparkContext.getOrCreate(SparkConf().setMaster('local[*]').setAppName('transformation_exercises'))

web_logs = sc.parallelize([
    "192.168.1.1 GET /home 200 150ms",
    "192.168.1.2 GET /products 200 230ms",
    "192.168.1.1 POST /login 200 180ms",
    "192.168.1.3 GET /home 404 50ms",
    "192.168.1.2 GET /products 200 210ms",
    "192.168.1.1 GET /home 200 120ms",
    "192.168.1.4 GET /admin 403 30ms"
])


split_logs = web_logs.map(lambda x: x.split())

successful_logs = split_logs.filter(lambda x: x[3] == "200")

endpoints = successful_logs.map(lambda x: x[2])

endpoint_pairs = endpoints.map(lambda x: (x, 1))

endpoint_counts = endpoint_pairs.reduceByKey(lambda a, b: a +b)

print(f"Successful Logs:  {successful_logs.collect()}")
print(f"Endpoints:  {endpoints.collect()}")
print(f"Endpoint counts:  {endpoint_counts.collect()}")


sc.stop()
```
