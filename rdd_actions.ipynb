```
from pyspark import SparkContext, SparkConf

conf = SparkConf().setMaster("local[*]").setAppName("RddActions")
sc = SparkContext.getOrCreate(conf)


numbers = sc.parallelize([10, 5, 8, 3, 15, 12, 7, 20, 1, 9])

# Task A:  all elements
print(f"Nums:  {numbers.collect()}")

# Task b:  count of elements

print(f"Count:  {numbers.count()}")

# Task c:  get first element

print(f"First:  {numbers.first()}")

# Task d:  get n elements
print(f"First 3:  {numbers.take(3)}")

#Task e:  largest n
print(f"Top 3:  {numbers.sortBy(lambda x: x, ascending=False).take(3)}")

# Task f:  smallest n
print(f"Bottom 3:  {numbers.sortBy(lambda x: x, ascending=True).take(3)}")


# Task 2.  
# 2a
print(f"Total:  {numbers.reduce(lambda x, y: x + y)}")

#2b
print(f"Max:  {numbers.reduce(lambda x, y: max(x,y))}")

#2c
print(f"Min:  {numbers.reduce(lambda x, y: min(x,y))}")

#2d
print(f"Folded sum:  {numbers.fold(0, lambda x, y: x + y)}")


#3a
colours = sc.parallelize(["red", "blue", "red", "green", "blue", "red", "yellow"])
colour_key_values = colours.map(lambda x: (x, 1))

colour_counts = colour_key_values.reduceByKey(lambda x, y: x+y).sortBy(lambda x: x[1], ascending=False)

print(f"Colour counts:  {colour_counts.collect()}")



sc.stop()
```
